```python
## generate dates between two dates

from datetime import date, timedelta

d1 = date(2018,1,1)  # start date
d2 = date(2019,4,1)  # end date
delta = d2 - d1         # timedelta

lst=[]
for i in range(delta.days + 1):
    date=d1+timedelta(i)
    date=int(date.strftime('%Y%m%d'))
    lst.append(date)
```
