```python
def stcp(stc,dat):
    ticker = stc
    yahoo_financials = YahooFinancials(ticker)
    
    
    d1 = dat  # start date
    start=d1-timedelta(5)
    end=d1+timedelta(5)
    start=start.strftime('%Y-%m-%d')
    end=end.strftime('%Y-%m-%d')
    
    
    hist_stock_prices = yahoo_financials.get_historical_price_data(start,end,'daily')
    hist_stock_prices
    
    df1 = pd.DataFrame(hist_stock_prices[stc]['prices'])
    df1['company'] = stc
    df1 = df1[['company']+list(df1.columns[:-1])]
    
    return df1
    
```

```python
data_final=stcp(data1.stocks[0],data1.date2[0])
```

```python
data_final1=data_final.copy()
data_final1
```

```python
for i in range(20)[1:]:
    try:
        data_final1=data_final1.append(stcp(data1.stocks[i],data1.date2[i]))
    except:
        None 
```

```python
data_final1.shape
```

```python
## selecting columns 

data_final2=data_final1[['company','formatted_date','high','low','open','volume']]
data_final2['formatted_date']=pd.to_datetime(data_final2['formatted_date']).dt.date
data_final2=data_final2.reset_index(drop=True)
data2 = data1[['stocks','date2','volume','percent_chg']]
data2= data2.rename(columns={'stocks': 'company','date2':'formatted_date'})
result = pd.merge(data_final2, data2, how='left', on=['company','formatted_date'])
```

## financial info for the stocks

```python
stk_stat={}
stk_list=list(result.company.unique())
for i in stk_list:
    try:
        yf = YahooFinancials(i)
        ksd=yf.get_key_statistics_data()
        stk_stat[list(ksd.keys())[0]]=list(ksd.values())[0]
    except:
        pass
    
stk_stat_df=pd.DataFrame.from_dict(stk_stat, orient='index').reset_index()

stk_stat_df=stk_stat_df.rename(columns={'index': 'company'})

stk_stat_df.head()
```
